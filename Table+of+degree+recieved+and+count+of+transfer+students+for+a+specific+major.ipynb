```python
#1.	This cell imports important libraries needed to run calculations. Data and data1 are variables assigned to the two different sheets in the excel raw data.
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline

from scipy import stats

import seaborn as sns

data=pd.read_excel('Transfer_MSU.xlsx')

data1=pd.read_excel('Transfer_MSU1.xlsx')
```

```python
#2.	This cell creates:
#1.	Major: a list of the majors from the second excel sheet.
#2.	Degree_conferred: a list of whether or not a degree was received from the second excel sheet.
#3.	TransferY_N: a list from the first excel sheet that contains whether or not a student transferred or not.
#4.	Years: list from the second sheet that contains the semester date of when they ended their major. (either switched or graduated)

major= []
for row in data1.values[:,2]:
    major.append(row)
    
    
degree_conferred = []
for row in data1.values[:,8]:
    degree_conferred.append(row)
    
    
transferY_N=[]
for row in data.values[:,2]:
    transferY_N.append(row)

years=[]
for row in data1.values[:,7]:
    years.append(row)
    
```

```python
#3.	This cell partitions the transferY_N list into two lists. One list is of the ID’s of students who transferred and the other is a list of ID’s of students who did not transfer. (ID_transfer_Y and ID_transfer_N)
count=0

ID_transfer_Y=[]

ID_transfer_N=[]

for i in transferY_N:
    
    if i=="Y":
        ID_transfer_Y.append(data.values[count,0])
        count+=1
   
    if i=="N":
        ID_transfer_N.append(data.values[count,0])
        count+=1
```

```python
#4.	This cell creates a function that inputs a major name and a list of years (05,14,etc) and outputs four lists. 
#1.	ID_grad: a list of ID’s of students who graduated in a certain time period(years list)
#2.	The other three lists are just the same ID’s but sorted out by which semester they graduated

#default is all years and physics
def ID_Grad_Major_Year(major_name='Physics',
                       years_list=["05","06","07","08","09","10","11","12","13","14","15","16"]):
    
    ID_grad=[]
    
    ID_grad_Summer=[]
    
    ID_grad_Fall=[]
    
    ID_grad_Spring=[]
    
    for a in years_list:
        count=0
        for i,j,k in zip(major,degree_conferred,years):

            if i==major_name and j=="Y" and k=="US"+a:

                ID_grad.append(data1.values[count,0])
                
                ID_grad_Summer.append(data1.values[count,0])

                count+=1
            
            elif i==major_name and j=="Y" and k=="FS"+a:

                ID_grad.append(data1.values[count,0])

                ID_grad_Fall.append(data1.values[count,0])
                
                count+=1
            
            elif i==major_name and j=="Y" and k=="SS"+a:

                ID_grad.append(data1.values[count,0])

                ID_grad_Spring.append(data1.values[count,0])
                
                count+=1


            else:

                count+=1


    return (ID_grad, ID_grad_Summer, ID_grad_Fall, ID_grad_Spring)
```

```python
#6.	This cell contains a list of majors that we found to be of interest.
major_list_1=["Physics","Human Biology","Physiology","Actuarial Science",
              "Chemistry","Computer Science","Mathematics","Astrophysics","Statistics","Zoology","Kinesiology"]
```

```python
#7.	This cell defines a function that takes in a list of transfer ID’s, non transfer ID’s, and a list of graduate ID’s and outputs a count of ID’s that transferred and graduated and a count of ID’s that did not transfer and did graduate.
def transfer_counter(list_of_transfers,list_of_non_transfers,list_of_grads):
    
    count_of_transfer_grad=0
    
    count_of_not_transfer_grad=0
    
    for i in list_of_transfers:
        
        for j in list_of_grads:
            
            if i==j:
                
                count_of_transfer_grad+=1
    
    for i in list_of_non_transfers:
        
        for j in list_of_grads:
            
            if i==j:
                
                count_of_not_transfer_grad+=1
                
    return(count_of_transfer_grad,count_of_not_transfer_grad)
```

```python
#8.	This cell defines a function that inputs a list of transfer students, list of non transfer students, a major list, and a list of years. It outputs a data frame that shows the count of transfer and non transfer graduates associated to a particular major. 
def dataframe_maker(list_transfer, list_not_transfer, major_list,years_list):
    
    transfer_count=[]
    
    not_transfer_count=[]
    
    for h in major_list:
        
        a,b,c,d=ID_Grad_Major_Year(h,years_list)
        
        e,f=transfer_counter(list_transfer,list_not_transfer,a)
       
        transfer_count.append(e)
        
        not_transfer_count.append(f)
    
    s=pd.DataFrame(np.column_stack([major_list,transfer_count,not_transfer_count]))
    
    s.columns = ['Degree Received Major', 'Count of Transfer','Count of Non-transfer']
    
    return s
```

```python
#9.	This cell calls the previous function
s=dataframe_maker(ID_transfer_Y,ID_transfer_N,major_list_1,years_list)
```
